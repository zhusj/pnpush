```python
def run_physics_engine(filename, id, mass, friction, force):

    location =[]
    rotation = []  
    
    time = 0
    print 'input:', g_blender_executable_path, blank_file, render_code, robot_file, model_file, time, \
    id, filename, force_gt, mass, friction, table_file, temp_dirname

    
    render_cmd = '%s %s -b --python %s -- %s %s %s %s %s %s %s %s %s %s' % \
    (g_blender_executable_path, blank_file, render_code, robot_file, model_file, time, \
    id, filename, force_gt, mass, friction, table_file, temp_dirname)

    p = os.popen(render_cmd,"r")
    while 1:
        line = p.readline()
        if not line: 
            print 'not line'
            break
#         if "filename" in line: 
#             print "filename:", line
        if "gt_loc" in line: 
            print "gt_loc:", line
        if "gt_rot" in line: 
            print "gt_loc:", line
        # if "frame" in line: 
        #     print "frame:", line
        if "location" in line: 
            print "output:", line
            vals = line.split(',')
            count = 0
            for val in vals:
                count += 1
                if count == 1:     
                    x= float(val[20:26])
                if count == 2:
                    y= float(val[-7:])
                if count == 3:    
                    z= float(val[:7])
        if "rotation" in line: 
            print "output:", line
            vals = line.split(',')
            count = 0
            for val in vals:
                count += 1
                if count == 1:
                    newstr = val[25:32].replace("=", "")
                    qw= float(newstr)
                if count == 2:
                    newstr = val[-7:].replace("=", "")
                    qx= float(newstr)
                if count == 3:
                    newstr = val[-7:].replace("=", "")
                    qy= float(newstr)
                if count == 4:
                    newstr = val[2:9].replace("=", "")
                    qz= float(newstr)
                    
    location.append([x, y])
    location = np.asarray(location)
    rotation.append([qw,qx,qy,qz])
    rotation = np.asarray(rotation)
    return location, rotation
```
